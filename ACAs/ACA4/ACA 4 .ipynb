```python
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Data Collection

After a total of 21 participants finished the survey, we export the survey result directly from the Tencent survey platform in a excel format.

#### Data Cleaning

In order to transform original excel sheet into an adjacency matrix, we directly use excel function to achieve it. Here is result after transformation.

```python
SNAdata = pd.read_csv("ACA4 data 2.csv", index_col = 0)
G = nx.Graph(SNAdata)
SNAdata
```

```python
nx.draw(G, with_labels = True)
```

#### Definition and Interpretation of the network

Definition: This is a network of friends that is connected through individual airlines choices/past flight experience. The nodes are 21 participants' name took part in the survey. The edges are drawn with the number of common airlines between each two edge.

Interpretation: We can assume participants' potential connection with one another, and their orientation towards certain airlines. The number of edges each node has represents the number of participants who have chosen the same airline. Thus, the more edges one node has, the greater diversity of airlines one took in relation to others, and the greater probability one will choose either airline in the future(less predictive of choosing a certain airline). On the other hand, the less edge one has, the more predictive of its airline preference.

```python
density = nx.density(G)

print('The edge density is: %.3f' % (density)) 
```

```python
degree = nx.degree(G)
degree
```

```python
pd.DataFrame(degree).sort_values(by=1,ascending=False)
```

From the above dataframe, we count the total number of intersection of airlines one has taken in relation to other participants in the survey. As we can see, Eva, Kkk, and Katie have the highest intersections with other participants in past airlines choices.

```python
close = nx.closeness_centrality(G)
pd.DataFrame.from_dict(close, orient = 'index').sort_values(by=0,ascending=False)
```

From the above dataframe, closeness centrality measures "how quickly" each node (participant) can travel to the rest of the graph. We can see that eva and kkk has the highest centrality.

```python
between = nx.betweenness_centrality(G)
pd.DataFrame.from_dict(between, orient = 'index').sort_values(by=0,ascending=False)
```

From the above dataframe, betweenness centrality shows the amount of influence a node has over the flow of information in a graph. We can find that kkk, as the node, serve as a bridge through the graph, as it has the shortest paths between all pairs of nodes.

Lets' see how many communities we can detect by using 'greedy_modularity_communities()'

```python
from networkx.algorithms.community.modularity_max import greedy_modularity_communities

c = list(greedy_modularity_communities(G))

print(len(c))
```

We only get 1 community with all the participants.

```python
community_0 = sorted(c[0])



print(community_0)
```

```python
np.random.seed(123)

# Obtain the positions of the nodes
pos = nx.spring_layout(G)

# Print out all the nodes with color
nx.draw_networkx_nodes(G, pos = pos, nodelist=community_0, node_color='yellow', label = True)


# Print out the edges
nx.draw_networkx_edges(G, pos = pos, width = 0.2)

# Make sure the label is printed
nx.draw_networkx_labels(G, pos=pos)

# Show the plot so it supress the other networkx outputs
plt.show()
```

```python
nx.draw_networkx_nodes(G, pos = pos, nodelist = dict(degree).keys(), node_size = [v * 50 for v in dict(degree).values()], label = True)
nx.draw_networkx_edges(G, pos = pos, width = 0.2)
nx.draw_networkx_labels(G, pos=pos)
plt.show()
```

Now we changed the edge depending on the weight it has. We can obtain the weight of each edge using 'get_edge_attributes()'. Also, we are making the width of the edge as 2**e/100 for clear representation of difference.



```python
eWeight = nx.get_edge_attributes(G,'weight')

nx.draw_networkx_nodes(G, pos = pos, nodelist = dict(degree).keys(), node_size = [v * 50 for v in dict(degree).values()], label = True)
nx.draw_networkx_edges(G,pos, width = [2**e/100 for e in eWeight.values()])
nx.draw_networkx_labels(G, pos=pos)

plt.show()
```

```python
# Degrees of each community
community_0_degree = {c: dict(degree)[c] for c in community_0}




# Print out the nodes with color and size
nx.draw_networkx_nodes(G, pos = pos, 
                       nodelist=community_0,
                       node_color='green',
                       node_size = [v * 50 for v in community_0_degree.values()],
                       label = True)




# Print out the edges with proper width
nx.draw_networkx_edges(G,pos, width = [2**e/50 for e in eWeight.values()])

# Make sure the label is printed
nx.draw_networkx_labels(G, pos=pos)

# Show the plot so it supress the other networkx outputs
plt.show()
```
